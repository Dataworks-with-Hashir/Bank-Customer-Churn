#Exploring demographic factors that influence user behavior

##Inorder to solve this statement we can consider the correlation test for a demographic variable i.e, Age of user and Number of order.

```python
UserDetails[["Age","Total Orders"]]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
stats.probplot(UserDetails["Age"], dist="norm", plot=plt)
plt.title('Q-Q Plot for Age')

plt.subplot(1, 2, 2)
stats.probplot(UserDetails["Total Orders"], dist="norm", plot=plt)
plt.title('Q-Q Plot for Orders')

plt.show()
```

Since the data above closely align with diagonal line it suggest normality.
so we can use pearson test as a correlation test for checking relation between them.

```python
from scipy.stats import pearsonr
```

```python
corr, p_value = pearsonr(UserDetails["Age"], UserDetails["Total Orders"])
print("Correlation:", corr, "P-value:", p_value)
```

the above result says that Correlation coefficient is not close to 1 and P value is not < 0.05 so it is not statistically significant which says that Orders are not correlated with age or Demographic factor have no influence on User behaviour


#analyzing the relationship between cooking sessions and user orders


```python
Session = CookingSessions.groupby("User ID")["Session ID"].count().reset_index()
Session
```

```python
UD = UserDetails[["User ID","Total Orders"]]
UD
```

```python
SO = pd.merge(Session,UD,left_on="User ID",right_on="User ID",how="left")
SO.rename(columns={"Session ID": "Total Sessions"}, inplace=True)
```

```python
plt.stackplot(SO["User ID"],SO["Total Sessions"],SO["Total Orders"],labels=["Session","orders"])
plt.legend(loc='upper left')
plt.show()
```

##the above graphs clearly indicate the relationship between the number of sessions and total orders for all users
